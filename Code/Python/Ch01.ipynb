# 第 1 章 概述

## 导入库

```python
import networkx as nx

from modules.tools import graph_info, map_color
```

## 读取karate数据

```python
karate = nx.read_graphml('../../Data/karate.graphml')
graph_info(karate)
```

## 绘制karate网络结构图

```python
pos = nx.spring_layout(karate, seed=42)
nx.draw(
    karate, pos,
    labels=nx.get_node_attributes(karate, 'label'), 
    node_color=map_color(karate), edge_color='gray' 
)
```

## 读取Koenigsberg数据

```python
Koenigsberg = nx.read_graphml('../../Data/Koenigsberg.graphml')
graph_info(Koenigsberg)
```

## 绘制Koenigsberg网络结构图

```python
# multiple edges
pos = nx.spring_layout(Koenigsberg, seed=42)
nx.draw(
    Koenigsberg, pos,
    labels=nx.get_node_attributes(Koenigsberg, 'name'),
    connectionstyle=[f'arc3,rad={0.3*(i-0.5)}' for i in range(4)],
    node_color=map_color(Koenigsberg, 'name'), edge_color='gray'
)
```
